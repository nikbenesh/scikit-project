Import the libraries required

```python
import sklearn # import sklearn for building the classifier
import matplotlib.pyplot as plt # import matplotlib since I am going to build some plots 
import pandas as pd
```

Set the data file name to be a constant

```python
DATAFILE = "A2Q2.csv"
```

Read the data from the file and build the density plots (for all features and then for each).

```python
df = pd.read_csv(DATAFILE, header=None)

df.columns = ['Feature ' + str(i) for i in range(len(df.columns))]
del df['Feature 0']
df.plot.density() # build the density plot for all features
plt.show()

# build density plots for each feature
for col in df.columns:
    df[col].plot.density()
    plt.xlabel(col)
    plt.show()
```

```python
x = range(len(df.columns))
plt.bar(x, height=[len(set(df[col])) for col in df.columns])
plt.xticks(x, [i+1 for i in range(len(df.columns))])
plt.ylabel('Range')
plt.xlabel('Features')
plt.show()
```
